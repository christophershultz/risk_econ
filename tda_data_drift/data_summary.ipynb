```python
import pandas as pd
import numpy as np
import os, pdb, sys, pickle

with open('results.pkl', 'rb') as f: 
    dct = pickle.load(f)
dct.keys()
```

```python
df = {}
for key in dct.keys(): 
    name = key.split('-')[0]
    sub = dct[key][['x1', 'x2', 'x3']]
    reg = sub.iloc[:1000]
    drf = sub.iloc[1000:]
    df[name] = {'reg': reg, 'drf': drf}
print(df.keys())
```

```python
ks, meanReg, meanDrf, minReg, minDrf, maxReg, maxDrf, varReg, varDrf = [], [], [], [], [], [], [], [], []

for k in df.keys(): 
    reg = df[k]['reg']
    drf = df[k]['drf']
    ks.append(k)
    meanReg.append(np.nanmean(reg['x1']))
    meanDrf.append(np.nanmean(drf['x1']))
    minReg.append(min(reg['x1']))
    minDrf.append(min(drf['x1']))
    maxReg.append(max(reg['x1']))
    maxDrf.append(max(drf['x1']))
    varReg.append(np.var(reg['x1']))
    varDrf.append(np.var(drf['x1']))

res = pd.DataFrame({'shock': ks, 'mean_base': meanReg, 'mean_shocked': meanDrf, 
                   'min_base': minReg, 'min_shocked': minDrf, 
                   'max_base': maxReg, 'max_shocked': maxDrf, 
                   'var_base': varReg, 'var_shocked': varDrf})
res
```
